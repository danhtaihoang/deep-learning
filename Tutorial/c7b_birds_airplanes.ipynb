```python
## we also can transforms the whole dataset directly
tensor_cifar10 = datasets.CIFAR10(data_path, train=True, download=False,
                          transform=transforms.ToTensor())
```

### Normalizing data

Normalize the dataset so that each channel has zero mean and unitary standard deviation.

```python
## stack all the tensors returned:
imgs = torch.stack([img_t for img_t, _ in tensor_cifar10], dim=3)
imgs.shape
```

```python
## compute the mean per channel:
## note: view(3,-1) keeps the 3 channels and merges all the remaining dimensions into 1.
imgs.view(3, -1).mean(dim=1)
```

```python
imgs.view(3, -1).std(dim=1)
```

```python
## normalize based on the above values
transforms.Normalize((0.4915, 0.4823, 0.4468), (0.2470, 0.2435, 0.2616))

transformed_cifar10 = datasets.CIFAR10(
    data_path, train=True, download=False,
    transform=transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.4915, 0.4823, 0.4468),
                             (0.2470, 0.2435, 0.2616))
    ]))
```

```python
transformed_cifar10_val = datasets.CIFAR10(
    data_path, train=False, download=False,
    transform=transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.4915, 0.4823, 0.4468),
                             (0.2470, 0.2435, 0.2616))
    ]))
```
