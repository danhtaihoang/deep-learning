### Load metadata file

```python
metadata_file = "metadata_liver_skin.csv"

df = pd.read_csv(metadata_file)
df
```

```python
metadata = df.loc[:,['Image_name','Cancer_type']]
metadata
```

### Create index for image and target

```python
metadata_dict = metadata.to_dict()
print(metadata_dict)
```

### Get image and target from a given index

```python
idx = 0

## file name without .png
#file_name0 = image_names[idx]
file_name0 = metadata_dict["Image_name"][idx]
print(file_name0)

## file name with .png
file_name = f"/{file_name0}.png"
print(file_name)

## image_path
folder_name = "image_data"
image_path = folder_name + file_name
print("image_path:",image_path)

## load image
image = Image.open(image_path)
print("image.size:",image.size)
```

```python
image
```

```python
target = metadata_dict["Cancer_type"][idx]
print("target:", target)
```

### Convert target to class index/label

```python
## list of unique class names
classes = metadata["Cancer_type"].unique()
print(classes)

## class to index
class_to_idx = {j: i for i, j in enumerate(classes)}
print(class_to_idx)

label = class_to_idx[target]
print("label:", label)
```
